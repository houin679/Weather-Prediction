# MACHINE LEARNING TRAINING

```python
# Trying to predict tomorrow maximum temperature using historical data
```

```python
core_weather["target"] = core_weather.shift(-1)["temp_max"] # pulls every row back one position
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:] # deletes the last row
```

```python
core_weather
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha =.1)
```

```python
predictors = ["precip","temp_max","temp_min"]
```

```python
train = core_weather.loc[:"2020-12-31"]
```

```python
test = core_weather.loc["2021-01-01":]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"],predictions) 
```

```python
predictions
```

```python
combined = pd.concat([test["target"],pd.Series(predictions,index = test.index)], axis = 1)
combined.columns = ["actual", "predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_
```

```python
def create_predictions(predictors,core_weather, reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"],predictions) 
    combined = pd.concat([test["target"],pd.Series(predictions,index = test.index)], axis = 1)
    combined.columns = ["actual", "predictions"]
    return error,combined
```
